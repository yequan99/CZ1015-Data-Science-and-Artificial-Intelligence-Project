```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import seaborn as sb
sb.set()
```

## order_df

from 'olist_orders_dataset.csv'

```python
order_csv = pd.read_csv('olist_orders_dataset.csv')
order_csv = order_csv.drop(columns=['order_status','order_approved_at', 'order_delivered_carrier_date'])
order_csv
```

```python
order_csv['order_time'] = pd.to_datetime(order_csv['order_purchase_timestamp'], format = '%d/%m/%Y %H:%M')
order_csv['actual_delivery_time'] = pd.to_datetime(order_csv['order_delivered_customer_date'], format = '%d/%m/%Y %H:%M')
order_csv['estimated_delivery_time'] = pd.to_datetime(order_csv['order_estimated_delivery_date'], format = '%d/%m/%Y %H:%M')
order_csv = order_csv.drop(columns = ['order_purchase_timestamp', 'order_delivered_customer_date', 'order_estimated_delivery_date'])
order_csv
```

```python
order_csv['actual'] = pd.DataFrame(order_csv['actual_delivery_time'].sub(order_csv['order_time'], axis=0))
order_csv['estimated'] = pd.DataFrame(order_csv['estimated_delivery_time'].sub(order_csv['order_time'], axis=0))
order_csv['actual_hours'] = pd.DataFrame(order_csv['actual'].astype('timedelta64[h]'))
order_csv['delivery_days'] = pd.DataFrame(order_csv['actual'].astype('timedelta64[D]'))
order_csv['estimated'] = pd.DataFrame(order_csv['estimated'].astype('timedelta64[h]'))
order_csv['time_diff'] = pd.DataFrame(order_csv['estimated'].sub(order_csv['actual_hours'], axis=0))

for index, row in order_csv.iterrows():
    order_csv.at[index, 'PurchasedYear'] = row["order_time"].year
order_csv['PurchasedYear'] = order_csv['PurchasedYear'].astype('int64')

order_csv
```

## customer_df

from 'olist_customers_dataset.csv'

```python
customer_df = pd.read_csv('olist_customers_dataset.csv')
customer_df
```

## products_df 

#### df contains translated product category names

from 'olist_products_dataset.csv' and 'product_category_name_translation.csv'


```python
products_dataset = pd.read_csv('olist_products_dataset.csv')
products_dataset = products_dataset.drop(columns=['product_name_lenght', 'product_weight_g',
                                                   'product_length_cm', 'product_height_cm', 'product_width_cm'])
products_dataset
```

```python
catname_translation = pd.read_csv('product_category_name_translation.csv')
products_df = pd.merge(products_dataset, catname_translation, how = "left", on = "product_category_name")
products_df = products_df.drop(columns = ['product_category_name'])
column_title = ['product_id','product_category_name_english','product_description_lenght','product_photos_qty']
products_df = products_df.reindex(columns = column_title)
products_df
```

## order_items_df 

from 'olist_order_items_dataset.csv'

```python
order_items_df = pd.read_csv('olist_order_items_dataset.csv')
order_items_df = order_items_df.drop(columns=['order_item_id','shipping_limit_date', 'seller_id'])
order_items_df
```

## payment_df

from 'olist_order_payments_dataset.csv'

```python
payment_df = pd.read_csv('olist_order_payments_dataset.csv')
payment_df
```

```python
payment_df2 = payment_df.groupby(['order_id'],as_index = False).agg({'payment_type':'first','payment_value' : 'sum', 'payment_installments' : 'max'})
payment_df2 = payment_df2[payment_df2.payment_type != 'not_defined']
payment_df2
```

## review_df

```python
review_df = pd.read_csv('olist_order_reviews_dataset.csv')
review_df.head()
```

```python
review_df = review_df.drop(columns = ['review_id', 'review_comment_title', 'review_comment_message', 'review_creation_date', 'review_answer_timestamp'])
```

## Merging products_df with order_items_df

Just to add category names and photo quantity into order_items_df

```python
product_order_df = pd.merge(order_items_df, products_df, how = "left", on = "product_id")
product_order_df
```

## Merging customer_df with order_csv 

```python
order_csv2 = pd.merge(order_csv,customer_df,how="left",on="customer_id")
order_csv2
```

## Merging payment_df2 with order_csv

```python
df = pd.merge(order_csv2,payment_df2,how="left",on="order_id")
df
```

## Getting the cleaned df with review_df

```python
final_df = pd.merge(product_order_df,df,how="left",on="order_id")
final_df
```

```python
final_df2 = pd.merge(final_df, review_df, how='left', on = 'order_id')
final_df2.head()
```

```python
clean_df = final_df2.drop(columns=['order_time',
       'actual_delivery_time', 'estimated_delivery_time',])
title = ['order_id', 'product_id', 'customer_id', 'customer_unique_id', 'product_category_name_english', 'product_description_lenght', 'product_photos_qty','actual_hours','delivery_days',
       'estimated', 'time_diff','price', 'freight_value','payment_type', 'payment_value',
       'payment_installments','PurchasedYear', 'review_score']
clean_df = clean_df.reindex(columns = title)
clean_df = clean_df.rename(columns={"product_description_lenght" : "product_description_length"})
```
